```pyspark
# Import the necessary libraries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```pyspark
# Read the data from HDFS.
data_df = spark.read.csv('/user/hadoop/eng_data_1gram/eng_1M_1gram.csv', header=True)
```

```pyspark
#  Print out the rows to get the shape of the dataframe
nrows = data_df.count()
print("There are", nrows, "rows in the data frame.")
```

```pyspark
# Determine the number of columns and print out the shape of the data frame.
ncols = len(data_df.columns)
data_shape = (nrows, ncols)
print("The shape (rows, columns) of the data frame is", data_shape, ".")
```

```pyspark
# Show the schema of the data frame.
data_df.printSchema()
```

```pyspark
# Show a sample of the data.
data_df.show()
```

```pyspark
# Q 4b: Create a new DataFrame from a query using Spark SQL, filtering to include only the rows where the token is "data" 
# and describe the new dataset

from pyspark.sql.functions import col

fltrd_data_df = data_df.filter(col("token") == "data")
fltrd_data_df.count()

# Alternate method:
#fltrd_data_df2 = data_df.filter(data_df.token == "data")
#fltrd_data_df2.count()
```

```pyspark
# Q4b cont'd: Describe the filtered data set.
fltrd_nrows = fltrd_data_df.count()
fltrd_ncols = len(fltrd_data_df.columns)
fltrd_data_shape = (fltrd_nrows, fltrd_ncols)
print("The shape (rows, columns) of the filtered data frame is", fltrd_data_shape, ".")
print("The shape (rows, columns) of the original data frame was", data_shape, ".")
```

```pyspark
# Q4c: Write the filtered data back to a directory in the HDFS from Spark
fltrd_data_df.write.csv('/user/hadoop/eng_fltrd_data_1gram/fltrd_eng_1M_1gram', header=True)
```
